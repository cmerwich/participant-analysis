```python
import sys, os, re, pickle, csv, collections
from collections import *
from IPython.display import HTML
from pprint import pprint
from functools import reduce

from tf.fabric import Fabric
from tf.transcription import Transcription
from tf.extra.bhsa import Bhsa
#from participant_helpers import * 

import numpy as np
from pandas import DataFrame, read_csv
import pandas as pd
from random import random
import matplotlib.pyplot as plt
```

```python
VERSION = '2017'
DATABASE = '~/github/etcbc'
BHSA = f'bhsa/tf/{VERSION}'
REFERENCE = f'bh-reference-system/tf/{VERSION}'
HEADS = f'lingo/heads/tf/{VERSION}'
TF = Fabric(locations=[DATABASE], modules=[BHSA, REFERENCE, HEADS], silent=False )
```

```python
api = TF.load('''
    otype
    lex book chapter verse
    nu ps gn prs ls lex gloss
    function sp typ pdp language 
    pgn_prps pgn_prde pgn_verb 
    pgn_verb_prs pgn_prs 
    heads prep_obj
''', silent=True)

api.makeAvailableIn(globals())
#F, E, T, L = api.F, api.E, api.T, api.L

B = Bhsa(api, 'coref_rule_based', version=VERSION)
```

```python
MY_BOOK = {'Psalms'}
MY_CHAPTERS = set(range(1,2)) 

results = []
highlights = {}

for book in F.otype.s('book'):
    book_name = T.bookName(book)
    
    for chn in L.d(book, 'chapter'):
        chapter = F.chapter.v(chn)
        tup = ()
        if book_name in MY_BOOK and chapter in MY_CHAPTERS:
            
            for phrase in L.d(chn, 'phrase'):
                heads = E.heads.f(phrase)
                bo, ch, vs = T.sectionFromNode(heads[0])
                tup = tup + (heads[0],)
                highlights[heads[0]] = 'tomato'
                
                if F.typ.v(phrase) == 'PP':
                    prep_objs = [E.prep_obj.f(prep)[0] for prep in heads if E.prep_obj.f(prep)]
                    if prep_objs:
                        tup = tup + (prep_objs[0],)
                
                        highlights[prep_objs[0]] = 'springgreen'
                
            results.append(tup)
            #print(results)
            
                    #print(prep_objs[0])
                    #print(bo, ch, vs, T.text(heads), T.text(prep_objs), prep_objs)
B.prettySetup(features='pgn_prps pgn_prde pgn_verb pgn_prs')           
B.show(results, condensed=False, highlights=highlights, withNodes=True)
```
