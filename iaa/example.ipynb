```python
import os
import pandas as pd
from retrieve_iaa import retrieve_ann
from acc import print_total
from utils import MakeTable 
```

```python
column_names=('ann_A','ann_B', 'L', 'M', 'R', 'D', 'd') 
data_types={'ann_A': str, 'ann_B': str ,'L': int, 'M': int, 'R': int, 'D': int, 'd': float}

def MakeTable(iaa_file):
    iaa_table = pd.read_table(iaa_file, 
                           delim_whitespace=True, 
                           names=column_names,
                           dtype=data_types
                          )
    cols = ['L', 'ann_A', 'M', 'ann_B', 'R', 'D', 'd']
    iaa_table = iaa_table[cols]
    decimals = 2    
    iaa_table['d'] = iaa_table['d'].apply(lambda x: round(x, decimals))
    
    return iaa_table
MakeTable('iaa-files/Psalms_138.iaa')
```
