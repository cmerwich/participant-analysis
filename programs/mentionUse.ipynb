```python
# cruft functions for checking data

def print_words(book, chapter):
    chn = T.nodeFromSection((book, chapter))
    mnodes = tuple(n for n in L.d(chn) if E.nlink.t(n))
    
    colors = [F.color.v(m) for m in F.otype.s('mention')]
    for n in mnodes:
        words = T.text(n, fmt='text-trans-plain')
        
    for m in F.otype.s('mention'):
        suf = F.mSuf.v(m)
        det = F.mDet.v(m)
        prs = F.prs.v(L.d(m, 'word')[0])
        color = F.color.v(m)
        if suf and prs:
            print(suf, prs, det, T.text(m, fmt='text-trans-plain'), color, sep='\t')
            
print_words('Psalms', 1)
```

```python
def transPrs(value):
    prs = int
    if value == 'absent':
        prs = 0
    elif value == 'n/a':
        prs = 0
    else:
        prs = 1
    return prs 
  
l=[]
ll=[]
def prs(book, chapter):
    chn = T.nodeFromSection((book, chapter))
    for wn in L.d(chn, 'word'):
        t = T.text(wn, fmt='text-trans-plain')
        p = F.prs.v(wn)
        prs = transPrs(F.prs.v(wn))
        if prs == 1:
            l.append(prs)
        if prs == 0:
            ll.append(prs)
        print(t, p, prs, sep='\t')
        
    print(len(l))
    print(len(ll))
        
prs('Psalms', 1)
```
